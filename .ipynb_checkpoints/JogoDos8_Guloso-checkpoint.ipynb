### Método jogar
A função jogar é a função principal do jogo, onde são calculados os possíveis movimentos de acordo com o estado atual e também, aonde é feita a escolha de qual desses movimentos será escolhido. 
Para fazer essa escolha temos a possibilidade de 3 tipos de criterios e a função escolher critério irá calcula-lo de acordo com o seu valor (mais a baixo).

O critério guarda é uma string responsável por evitar movimentos ciclicos (por exempo, cima-baixo-cima ou esquerda-direita-esquerda). Além disso, o algoritmo está limitado a 200 iterações para evitar o desgaste de recursos ao entrar em loop infinito.

```python
def jogar(atual, final, guarda, criterio):
    count = 0

    matriz_baixo = numpy.zeros(shape=(3,3))
    matriz_cima = numpy.zeros(shape=(3,3))
    matriz_esquerda = numpy.zeros(shape=(3,3))
    matriz_direita = numpy.zeros(shape=(3,3))

    inicio = time.time()
    while count < 200:
        crit_cima = 99999
        crit_direita = 99999
        crit_esquerda = 99999
        crit_baixo = 99999

        zero_atual = encontra(0, atual)
        posicaoi = zero_atual[0]
        posicaoj = zero_atual[1]

        crit_atual = escolherCriterio(criterio, atual, final)

        if crit_atual == 0:
            fim = time.time()
            #print("O jogo foi encerrado na "+ str(count)+" iteração com "+ str((fim-inicio)) +" segundos")
            return (fim-inicio), count

        if posicaoi - 1 >= 0:
            matriz_cima = atual.copy()
            matriz_cima = trocarPosicoes(matriz_cima, 4, posicaoi, posicaoj)
            crit_cima = escolherCriterio(criterio, matriz_cima, final)
        if posicaoi + 1 < 3:
            matriz_baixo = atual.copy()
            matriz_baixo = trocarPosicoes(matriz_baixo, 3, posicaoi, posicaoj)
            crit_baixo = escolherCriterio(criterio, matriz_baixo, final)
        if posicaoj - 1 >= 0:
            matriz_esquerda = atual.copy()
            matriz_esquerda = trocarPosicoes(matriz_esquerda, 1, posicaoi, posicaoj)
            crit_esquerda = escolherCriterio(criterio, matriz_esquerda, final)
        if posicaoj + 1 < 3:
            matriz_direita = atual.copy()
            matriz_direita = trocarPosicoes(matriz_direita, 2, posicaoi, posicaoj)
            crit_direita = escolherCriterio(criterio, matriz_direita, final)

        if crit_cima<=crit_baixo and crit_cima<=crit_esquerda and crit_cima<=crit_direita and crit_cima!=99999 and crit_cima<=crit_atual:
            guarda_temp = "cima"
            if guarda == "baixo":
                crit_cima = 99999;
            else:
                atual = matriz_cima.copy()
        if crit_baixo<=crit_cima and crit_baixo<=crit_esquerda and crit_baixo<=crit_direita and crit_baixo!=99999 and crit_baixo<=crit_atual:
            guarda_temp = "baixo"
            if guarda == "cima":
                crit_baixo = 99999;
            else:
                atual = matriz_baixo.copy()
        if crit_direita<=crit_baixo and crit_direita<=crit_esquerda and crit_direita<=crit_cima and crit_direita!=99999 and crit_direita<=crit_atual:
            guarda_temp = "direita"
            if guarda == "esquerda":
                crit_direita = 99999;
            else:
                atual = matriz_direita.copy()
        if crit_esquerda<=crit_baixo and crit_esquerda<=crit_direita and crit_esquerda<=crit_cima and crit_esquerda!=9999 and crit_esquerda<=crit_atual:
            guarda_temp = "esquerda"
            if guarda == "direita":
                crit_esquerda = 99999;
            else:
                atual = matriz_esquerda.copy()

        crit_atual = escolherCriterio(criterio, atual, final)
    
        guarda = guarda_temp
        count = count + 1
```

### Método escolher critério: escolhe o critério em que o jogo vai se basear para encontrar o estado final da matriz. Os três primeiros métodos são métodos informados, e o quarto é o método desinformado.

```python
def escolherCriterio(criterio, atual, final):
    if criterio == 1:
        crit_atual = calcularDistanciaEuclidiana(atual, final)
    elif criterio == 2:
        crit_atual = calcularPosicoesDiferentes(atual, final)
    elif criterio == 3:
        crit_atual = calcularMd(atual, final)
    elif criterio == 4:
        crit_atual = buscaEmLargura(atual, final)
            
    return crit_atual
```

### Método calcularDistanciaEuclidiana

Calcula a distância euclidiana, para a matriz, ou seja, a soma da distância euclidiana de todos os elementos de 1 a 8

```python
def calcularDistanciaEuclidiana(estado_atual, estado_final):
    euclid = 0
    
    for e in range(1,9):
        pos_atual = encontra(e, estado_atual)
        x1=pos_atual[0]
        y1=pos_atual[1]
        pos_final = encontra(e, estado_final)
        x2=pos_final[0]
        y2=pos_final[1]
        e_euclid = ((((x1-x2)**2)+((y1-y2)**2))**0.5)
        euclid = euclid + e_euclid
        
    return euclid
```

### Metódo calcularPosicoesDiferentes

Calcula quantos elementos na matriz estão em posições diferentes da sua posição objetivo

```python
def calcularPosicoesDiferentes(estado_atual, estado_final):
    dif = 0
    
    for a in range(0,3):
        for b in range(0,3):
            if estado_atual[a,b] != estado_final[a,b] and estado_atual[a,b]!=0:
                dif = dif+1
                
    return dif
```

### Metódo calcularMd

Calcula a Manhattan distance para a matriz toda, ou seja, o somatório de todas as manhattan distance dos elementos de 1 a 8

```python
def calcularMd(estado_atual, estado_final):
    md = 0
    for x in range(1,9):
        mdi = 0
        posicao_atual = encontra(x, estado_atual)
        posicao_final = encontra(x, estado_final)
        i = posicao_atual[0]
        j = posicao_atual[1]
        igual = bool(0)
        while not igual:
            if i!=posicao_final[0] : 
                mdi = mdi+1
                if i >posicao_final[0]:
                    i = i-1
                else:
                    i = i+1
            elif j!=posicao_final[1]:
                mdi = mdi+1
                if j>posicao_final[1]:
                    j = j-1
                else:
                    j = j+1
            else:
                igual = bool(1)
        md = md + mdi
    return md
```

### Método Trocar Posições: 
Troca entre o elemento 0, ou seja, o elemento vazio, e um movimento conforme indicado pela variável lado:


*   lado = 1: troca com o elemento da esquerda
*   lado = 2: troca com o elemento da direita
*   lado = 3: troca com o elemento de baixo
*   lado = 4: troca com o elemento de cima



```python
def trocarPosicoes(m, lado, posicaoi, posicaoj):
    
    if lado == 1: #esquerda
        novo_j=posicaoj -1
        aux = m[posicaoi, novo_j]
        aux2 = m[posicaoi, posicaoj]
        m[posicaoi, posicaoj] = aux
        m[posicaoi, novo_j] = aux2
    elif lado == 2: #direita
        novo_j=posicaoj + 1
        aux = m[posicaoi, novo_j]
        aux2 = m[posicaoi, posicaoj]
        m[posicaoi, posicaoj] = aux
        m[posicaoi, novo_j] = aux2
    elif lado == 3: #baixo
        novo_i=posicaoi + 1
        aux = m[novo_i, posicaoj]
        aux2 = m[posicaoi, posicaoj]
        m[posicaoi, posicaoj] = aux
        m[novo_i, posicaoj] = aux2
    elif lado == 4: #cima
        novo_i = posicaoi - 1
        aux = m[novo_i, posicaoj]
        aux2 = m[posicaoi, posicaoj]
        m[posicaoi, posicaoj] = aux
        m[novo_i, posicaoj] = aux2
        
    return m
```

### Método escreverArquivo

Escreve no arquivo de acordo com o formato:

"tempo", "iterações"

E se passar acabou=true, ele adiciona "================================" o que indica que os tempos a frente são para uma matriz diferente

```python
def escreverArquivo(temp, count, acabou):
    arquivo = open("resultados_guloso.txt", "a")
    arquivo.write(str(temp))
    arquivo.write(",")
    arquivo.write(str(count))
    arquivo.write("\n")
    if acabou: arquivo.write("================================ \n")
```

### Método Encontra: 
Encontra um elemento em uma matriz

```python
def encontra(x, estado_atual):
    for i in range(3):
        for j in range(3):
            if x == estado_atual[i,j]:
                return i,j
```

### Metódo executarDiferentesCriteriosEEscreverArquivo

Para um estado atual e um estado objetivo, a função roda todos os critérios diferentes e os escreve no arquivo. Além disso, printa em um texto de alto nível os resultados obtidos



```python
def executarDiferentesCriteriosEEscreverArquivo(matriz_atual, matriz_objetivo):
    ultimo = "" 
    temp_euc, count_euc = jogar(matriz_atual, matriz_objetivo, ultimo, 1)
    temp_dif, count_dif = jogar(matriz_atual, matriz_objetivo, ultimo, 2)
    temp_md, count_md = jogar(matriz_atual, matriz_objetivo, ultimo, 3)

    escreverArquivo(temp_euc, count_euc, bool(0))
    escreverArquivo(temp_dif, count_dif, bool(0))
    escreverArquivo(temp_md, count_md, bool(1))

    print("Euclidiana: O jogo foi encerrado na "+ str(count_euc) +" iteração com "+ str(temp_euc) +" segundos")
    print("Posições Diferentes: O jogo foi encerrado na "+ str(count_dif) +" iteração com "+ str(temp_dif) +" segundos")
    print("MD: O jogo foi encerrado na "+ str(count_md) +" iteração com "+ str(temp_md) +" segundos")
    print("\n")
```
